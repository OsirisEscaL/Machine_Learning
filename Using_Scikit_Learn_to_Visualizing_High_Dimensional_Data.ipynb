# Using Scikit-Learn to Visualizing High-Dimensional Data

It is common in data science to encounter high-dimensional datasets that challenge our ability to visualize and comprehend the underlying patterns. Principal Component Analysis (PCA) and other dimensionality reduction techniques save the day. This article will examine how to visualize high-dimensional data in 2D or 3D space using PCA and scikit-learn.

**Dataset**

We require a data set with pertinent characteristics and household values. This dataset is available in [Kaggle](https://www.kaggle.com/datasets/fedesoriano/the-boston-houseprice-data). Now you have a CSV file named 'house_prices_dataset.csv' containing the necessary data for this example.

**Step 1: Data Preprocessing**

Importing the essential Python libraries for the project will be our initial step:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

An essential step in prepping a dataset for modeling is data preprocessing. It includes data entry, modification, and transformation.

```python
# Load your dataset
data = pd.read_csv('house_prices_dataset.csv')
# Check the first few rows of the dataset
data.head()
```

```python
# Separate features (X) and target variable (y)
X = data.drop('MEDV', axis=1)
y = data['MEDV']
```

PCA is sensitive to the scale of the data, so standardizing the features is essential:

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Step 2: Apply PCA**

Principal Component Analysis, or PCA, is an extensively employed dimensionality reduction method in data analysis, machine learning, and statistics. We predominantly used to reduce the dimensionality of high-dimensional data while preserving the most essential information and patterns. PCA accomplishes this by transforming the initial correlated characteristics into a new set of orthogonal (uncorrelated) features known as principal components.

We create a PCA object and indicate the number of retained components. To display the data in two dimensions, set n_components to 2. To enable 3D visualisation, enter 3:

```python
n_components = 2

pca = PCA(n_components=n_components)
X_pca_2d = pca.fit_transform(X_scaled)
```

```python
n_components = 3

pca = PCA(n_components=n_components)
X_pca_3d = pca.fit_transform(X_scaled)
```

**Step 3: Visualize the Reduced Data**

Now that we have reduced the dimensionality of our dataset, let's visualize it using scatter plots. For 2D representation:

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of House Prices Dataset')
plt.colorbar(label='MEDV - Median value of owner')
plt.show()
```

For 3D visualization:

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca_3d[:, 0], X_pca[:, 1], X_pca_3d[:, 2], c=y, cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.title('PCA of House Prices Dataset')
plt.colorbar(scatter, label='MEDV - Median value of owner')
plt.show()
```

**Conclusion**

Combining PCA with scikit-learn and real-world datasets from platforms like Kaggle enables data scientists and analysts to discover latent patterns and insights in high-dimensional data. By following the methods outlined in this article, you can effectively reduce the dimensionality of your data and create insightful visualizations, allowing you to discover valuable patterns and relationships within your dataset.
